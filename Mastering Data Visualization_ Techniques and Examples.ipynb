# **Mastering Data Visualization: Techniques and Examples**

1.Bar Chart


```python
import matplotlib.pyplot as plt

# Sample Data
user_types = ['New Users', 'Returning Users', 'Guests']
number_of_purchases = [300, 800, 200]

# Create Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(user_types, number_of_purchases, color=['blue', 'green', 'orange'])

# Adding titles and labels
plt.title('Number of Purchases by User Type')
plt.xlabel('User Type')
plt.ylabel('Number of Purchases')

# Show the plot
plt.show()
```

2.Line Chart

```python
import matplotlib.pyplot as plt

# Sample Data
months = ['January', 'February', 'March', 'April', 'May', 'June']
sales = [5000, 7000, 6000, 8000, 9000, 12000]

# Create Line Chart
plt.figure(figsize=(8, 5))
plt.plot(months, sales, marker='o', color='blue')

# Adding titles and labels
plt.title('Monthly Sales Figures')
plt.xlabel('Month')
plt.ylabel('Sales ($)')

# Show the plot
plt.grid()
plt.show()
```

3.Pie Chart

```python
import matplotlib.pyplot as plt

# Sample Data
brands = ['Brand A', 'Brand B', 'Brand C', 'Brand D']
market_share = [40, 30, 20, 10]

# Create Pie Chart
plt.figure(figsize=(8, 5))
plt.pie(market_share, labels=brands, autopct='%1.1f%%', startangle=140)

# Adding title
plt.title('Market Share by Brand')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

4.Scatter Plot

```python
import matplotlib.pyplot as plt

# Sample Data
advertising_spend = [1000, 2000, 3000, 4000, 5000]
sales_revenue = [5000, 7000, 9000, 12000, 15000]

# Create Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(advertising_spend, sales_revenue, color='blue')

# Adding titles and labels
plt.title('Advertising Spend vs Sales Revenue')
plt.xlabel('Advertising Spend ($)')
plt.ylabel('Sales Revenue ($)')

# Show the plot
plt.grid()
plt.show()
```

5.Stacked Chart


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data: Sales data for different product categories over three quarters.
quarters = ['Q1', 'Q2', 'Q3']
electronics = [500, 700, 800]
clothing = [300, 400, 500]
home_goods = [200, 300, 400]

# Create Stacked Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(quarters, electronics, label='Electronics', color='blue')
plt.bar(quarters, clothing, bottom=electronics,
        label='Clothing', color='orange')
plt.bar(quarters,
        home_goods,
        bottom=np.array(electronics) + np.array(clothing),
        label='Home Goods',
        color='green')

# Adding titles and labels
plt.title('Sales by Product Category Over Quarters')
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.legend()

# Show the plot
plt.show()
```

6.Dual Axis Chart

```python
import matplotlib.pyplot as plt

# Sample Data: Monthly sales and profit figures.
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [5000, 7000, 6000, 8000, 9000, 12000]
profit = [1000, 1500, 1200, 1800, 2100, 3000]

fig, ax1 = plt.subplots()

# Create bar chart for sales
ax1.bar(months,
         sales,
         color='b',
         alpha=0.6,
         label='Sales')
ax1.set_xlabel('Month')
ax1.set_ylabel('Sales ($)', color='b')

# Create second y-axis for profit
ax2 = ax1.twinx()
ax2.plot(months,
         profit,
         color='r',
         marker='o',
         label='Profit')
ax2.set_ylabel('Profit ($)', color='r')

# Adding titles and legends
plt.title('Monthly Sales and Profit')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

7.Heat Map

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample Data: Randomly generated website traffic data for illustration
data = np.random.randint(1, 100, size=(24, 5))

# Create Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data, cmap='YlGnBu', annot=True)

# Adding titles and labels
plt.title('Website Traffic Heatmap (Hour vs Day)')
plt.xlabel('Day of the Week')
plt.ylabel('Hour of the Day')

# Show the plot
plt.show()
```
